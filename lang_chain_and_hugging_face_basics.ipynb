```python
from langchain import PromptTemplate, LLMChain

question="Who won the Cricket World Cup in the year 2011?"
template = """Question: {question}
Answer: Let's think step by step."""
prompt = PromptTemplate(template=template, input_variables=["question"])
```

```python
prompt
```

```python
llm_chain=LLMChain(llm=llm,prompt=prompt)
print(llm_chain.invoke(question))
```

```python
# to print only the answer
print(llm_chain.invoke(question)['text'])
```

HuggingFacePipeline

```python
from langchain_huggingface import HuggingFacePipeline
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
```

```python
# we cannot load all models locally so we use small llm like gpt2 to run locally

model_id="gpt2" # free model in hugging face by open ai

model=AutoModelForCausalLM.from_pretrained(model_id) # downloads gpt2
tokenizer=AutoTokenizer.from_pretrained(model_id) # downloads tokenizer for gpt2 >> Every model needs a tokenizer to talk to it
```

```python
# pipeline(first param is task to perform we get it from documentation, last token is size of ans)
# we build pipeline to talk model in local machines like the way we talk to model through api
pipe = pipeline("text-generation",model=model,tokenizer=tokenizer,max_new_tokens=100)
hf = HuggingFacePipeline(pipeline=pipe)
```

```python
hf
```

```python
print(hf.invoke("What is machine learning")) # we used print so the ans is more organised
```

```python
## Use HuggingfacePipelines With Gpu

gpu_llm = HuggingFacePipeline.from_model_id(
    model_id="gpt2",
    task="text-generation",
    device=0,  # replace with device_map="auto". # device = 0 means gpu and device = -1 is cpu
    pipeline_kwargs={"max_new_tokens": 100},
)
```

```python
from langchain_core.prompts import PromptTemplate

template = """Question: {question}

Answer: Let's think step by step."""
prompt = PromptTemplate.from_template(template)
```

```python
chain=prompt|gpu_llm


question="What is artificial intelligence?"
chain.invoke({"question":question})
```
